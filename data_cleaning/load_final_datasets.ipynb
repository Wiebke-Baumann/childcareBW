```python
# for each gridcell, calculate the number of children in kindergarten-age (nr_children_kiga_age) per number of kindergartens in reach of this gridcell (nr_kiga_in_reach)
# This way, we later only count every child once!
for index, row in tqdm(g.iterrows(), total = len(g)):

    if row["nr_kiga_in_reach"] == 0:
        continue
    else:
        g.loc[index, "nr_children_kiga_age/nr_kiga_in_reach"] = g.loc[index, "nr_children_kiga_age"]/g.loc[index, "nr_kiga_in_reach"]
```

```python
g.head()
```

```python
# calculate the amount of children for each kindergarten (sum_children_kiga_age_new)
# We therefore sum of the column "nr_children_kiga_age/nr_kiga_in_reach" of all grid cells that lie in the buffer of a kindergarten

def childsum_per_kiga(dfa, dfb):
    df1 = dfa.copy() # kindergartens
    df2 = dfb.copy() # gridcells

    df1["sum_children_kiga_age_new"] = 0
    df1["occupancy_rate"] = 0

    for index, row in tqdm(df1.iterrows(), total = len(df1)):

        # calculate the amount of children for each kindergarten (sum_children_kiga_age_new)
        for i in range(len(row["grid_cell_IDs_in_buffer"])):

                id = row["grid_cell_IDs_in_buffer"][i]
                
                nr_children = (df2.loc[df2['ID_1km'] == id]["nr_children_kiga_age/nr_kiga_in_reach"])

                df1.loc[index, "sum_children_kiga_age_new"] = (df1.loc[index, "sum_children_kiga_age_new"]) + int(nr_children)

        # calculate the occupancy rate for each kindergarten --> divided by 51 (average number of kindergarten places)
        df1.loc[index, "occupancy_rate"] = df1.loc[index, "sum_children_kiga_age_new"]/51

    return df1, df2
```

```python
k_term, g_term = childsum_per_kiga(k, g)
```

```python
# calculate the average of the kindergarten occupation rate for each gridcell
def coverage_index(dfa, dfb):
    df1 = dfa.copy() # kindergartens
    df2 = dfb.copy() # gridcells

    df1 = df1.drop_duplicates(subset = ["id_kindergarten"])

    df2["avg_kiga_occupation"] = 0
    df2["childcare_index"] = None
    df2["sum_occupancy"] = 0

    for index1, row1 in tqdm(df2.iterrows(), total = len(df2)):

        if (row1["total_population"] == 0): 
            # 0 = no population in this grid
            df2.loc[index1,"childcare_index"] = 0
            continue

        if ((row1["nr_kiga_in_reach"] == 0) and ((row1["total_population"] == 0) or (row1["nr_children_kiga_age"] == 0))):
            # 0 = no kindergartens in reach & no children (or no population)
            df2.loc[index1,"childcare_index"] = 0
            continue

        if ((row1["nr_children_kiga_age"] > 0) and (row1["nr_kiga_in_reach"] == 0)):
            # 1 = no coverage (children in the grid-cell, but no kindergarten in reach)
            df2.loc[index1, "childcare_index"] = 1
            continue
        
        # calculate the average kindergarten occupancy rate
        for i in range(len(row1["kiga_IDs_in_reach"])):

            id = row1["kiga_IDs_in_reach"][i]
            
            df2.loc[index1, "sum_occupancy"] += float(df1.loc[df1['id_kindergarten'] == id]["occupancy_rate"]) # wird immer als Serie zurückgegeben! Warum? -> keine doppelten ids

        df2.loc[index1, "avg_kiga_occupation"] = float(df2.loc[index1, "sum_occupancy"]) / row1["nr_kiga_in_reach"]

        # 2 = slightly insufficient coverage (kindergarten in reach, but too many children for the kindergarten-places available in the buffers of the kindergartens in reach)
        # 1 < avg_kiga_occupation < 1.2 --> ca. 10 children more than the kindergarten can occupy
        if ((row1["nr_kiga_in_reach"] > 0) and (df2.loc[index1, "avg_kiga_occupation"] > 1)):
            df2.loc[index1, "childcare_index"] = 2

        # 3 = sufficient coverage (1 kindergarten in reach & enough kindergarten places available in the buffer)
        if ((row1["nr_kiga_in_reach"] == 1) and (df2.loc[index1, "avg_kiga_occupation"] <= 1)):
            df2.loc[index1, "childcare_index"] = 3

        #4 = good coverage (multiple kindergartens in reach & enough kindergarten places available in the buffer )
        if ((row1["nr_kiga_in_reach"] > 1) and (df2.loc[index1, "avg_kiga_occupation"] <= 1)):
            df2.loc[index1, "childcare_index"] = 4

    return df1, df2
```

```python
for index, row in tqdm(g_index.iterrows(), total = len(g_index)):
    # medium insufficient coverage
    # 1.2 < avg_kiga_occupation < 2.0 --> ca. double the amount of children than the kindergarten can occupy
    if (row["childcare_index"] == 2) and (row["avg_kiga_occupation"] >= 1.2) and (row["avg_kiga_occupation"] < 2.0):
        g_index.loc[index, "childcare_index"] = 5

    # strong insufficient coverage
    # avg_kiga_occupation > 2.0 --> more than double the amount of children than the kindergarten can occupy
    if (row["childcare_index"] == 2) and (row["avg_kiga_occupation"] > 2.0):
        g_index.loc[index, "childcare_index"] = 6
```

```python
# rearrange the index values so it is a locigal scale (the higher the score, the better the kindergarten supply in each grid cell):

#   0 = no kindergartens in reach & no children (or no population)
#   1 = no kindergarten coverage (children in the grid-cell, but no kindergarten in reach)
#   2 = strongly overcrowded kindergartens (more than double the amount of children than the kindergartens in reach can occupy)
#   3 = medium overcrowded kindergartens (ca. double the amount of children than the kindergartens in reach can occupy)
#   4 = slightly overcrowded kindergartens (ca. 10 children more than the kindergartens in reach can occupy)
#   5 = sufficient kindergarten supply (1 kindergarten in reach & enough kindergarten places available in the buffer)
#   6 = good kindergarten supply (multiple kindergartens in reach & enough kindergarten places available in the buffer)

g = g_index.copy()

for index, row in tqdm(g.iterrows(), total = len(g)):
    if row["childcare_index"] == 2:
        g.loc[index, "childcare_index"] = 4

    if row["childcare_index"] == 3:
        g.loc[index, "childcare_index"] = 5

    if row["childcare_index"] == 4:
        g.loc[index, "childcare_index"] = 6

    if row["childcare_index"] == 5:
        g.loc[index, "childcare_index"] = 3

    if row["childcare_index"] == 6:
        g.loc[index, "childcare_index"] = 2
 
```
